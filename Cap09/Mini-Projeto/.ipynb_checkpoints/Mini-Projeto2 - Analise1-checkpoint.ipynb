## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
df.head()
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig = plt.figure(figsize = (12,8))
sns.distplot(df.yearOfRegistration, color = 'g')
plt.xlabel('Ano de Registro', fontsize = 15)
plt.ylabel('Densidade (KDE)', fontsize = 15)
plt.title('Distribuição de Veículos com Base no Ano de Registro', fontsize = 20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution-carla.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
fig1 = plt.figure(figsize = (12,8))
sns.boxplot(df['vehicleType'], df['price'])
plt.title("Análise de Outliers", fontsize = 20, color = 'r', pad = -30, horizontalalignment = 'left')
plt.xlabel('Tipo de Veículo', fontsize = 15)
plt.ylabel('Range de Preço', fontsize = 15)
plt.grid(axis = 'y')
plt.show()
```

```python
# Salvando o plot
fig1.savefig("plots/Analise1/price-vehicleType-boxplot-carla.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
#help(sns.countplot)
help(sns.set_palette)
```

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria
num = len(df['vehicleType'].unique())
listaHSV = [((1- x*1.0/num), 0.6, 0.9) for x in range(num)]

fig2 = plt.figure(figsize = (12,8))
ax = sns.countplot(df['vehicleType'], palette = listaHSV, saturation = 0.6)
plt.xlabel('Tipo de Veículo', fontsize = 15)
plt.ylabel('Total de Veículos para Venda', fontsize = 15)
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.2, p.get_height()+50))
plt.title('Contagem Total de Veículos à Venda conforme o Tipo de Veículo', fontsize = 20)
plt.show
```

```python
# gráfico alternativo
sns.set_style('white')
fig3 = sns.factorplot(x='vehicleType', data=df, kind='count', palette='RdPu')
fig3.fig.suptitle('Contagem Total de Veículos à Venda conforme o Tipo de Veículo', fontsize = 20)
fig3.fig.set_size_inches(15,8)
fig3.set_xlabels('Tipo de Veículo', fontsize = 15)
fig3.set_ylabels('Total de Veículos para Venda', fontsize = 15)
for p in fig3.ax.patches:
    fig3.ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+50))
```

```python
# Salvando o plot
fig3.savefig("plots/Analise1/count-vehicleType-carla.png")
```
