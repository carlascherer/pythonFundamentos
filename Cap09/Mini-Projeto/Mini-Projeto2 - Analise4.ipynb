```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
# Calcule a média de preço por marca e por veículo
df2 = pd.DataFrame()
for i in df['brand'].unique():
    for j in df['vehicleType'].unique():
        media = df[(df['brand'] == i) & (df['vehicleType'] == j)]['price'].mean()
        df2 = df2.append(pd.DataFrame({'brand': i, 'vehicleType': j, 'avgPrice': media}, index=[0]))

df2.avgPrice.fillna(0, inplace=True)
df2.index = range(0,len(df2))
df2.avgPrice.isnull().value_counts()
df2
```

```python
df2['avgPrice'] = df2['avgPrice'].astype(int)
df2
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
heat = df2.pivot('brand', 'vehicleType', 'avgPrice')
fig = sns.heatmap(heat, annot=True, linewidths=1.5, linecolor='white', xticklabels=True, yticklabels=True, fmt='d', cmap='YlGnBu')
fig.figure.set_size_inches(17,20)
fig.set_xlabel('Tipo de Veículo', fontsize=15)
fig.set_ylabel('Marca', fontsize=15)
fig.set_title('Heatmap - Preço médio de um veículo por marca e tipo de veículo', fontsize=20)
```

```python
# Salvando o plot
fig.figure.savefig("plots/Analise4/heatmap-price-brand-vehicleType-carla.png")
```
